## UNIVARIATE ANALYSIS :-

### **Histplot, Lineplot & Displot used for Continuous features.
### **Countplot used for Categorical & Discrete data.

## DEPARTMENT WISE PERFORMANCE RATING VISUALIZATION :-

```python
df = data[['EmpDepartment','PerformanceRating']]
```

```python
df
```

```python
df1 = pd.crosstab(df.EmpDepartment,df.PerformanceRating)
```

```python
df1.reset_index(inplace=True)
```

```python
df1.head()
```

```python
x = np.arange(len(df1['EmpDepartment']))
print(x)
```

```python
rcParams['figure.figsize'] = 20,10
plt.xticks(x,df1['EmpDepartment'])
plt.bar(x-0.2, df1[2],width=0.5,label='Good', color='Red')
plt.bar(x, df1[3],width=0.5,label='Excellent',color='Yellow')
plt.bar(x+0.2, df1[4],width=0.5,label='Outstanding',color='Pink')
plt.legend(loc='best',shadow=True)
plt.xlabel('Department',fontsize=15)
plt.ylabel('Number of people',fontsize=20)
plt.title('Department Wise Performance',fontsize=20)
plt.show()
```
