It displays the frequency or count of data points within predefined intervals (bins) as bars, providing insight into the data's underlying distribution and patterns.

## CORRELATION :-
### 1)  Correlation is a statistical measure that quantifies the degree to which two or more variables are related or associated. It assesses the strength and direction of the relationship between variables.
### 2) Common correlation coefficients include Pearson's correlation coefficient (for linear relationships), Spearman's rank correlation coefficient (for monotonic relationships), and others. 
### 3) A correlation value typically ranges from -1 (perfect negative correlation) to 1 (perfect positive correlation), with 0 indicating no linear correlation. Correlation does not imply causation; it only indicates a statistical association between variables.

```python
data.corr()
```

### For Plotting Correlation we use HeatMap.
## HEAT MAP :-
### 1) A heat map is a graphical representation of data where values in a matrix are represented as colors. It is often used to visualize the magnitude of a variable in a two-dimensional dataset.
### 2) In a typical heat map, each cell in the matrix is assigned a color shades based on the value it represents.
### 3) Heat maps are useful for identifying patterns, trends, and correlations in complex datasets, especially when dealing with large amounts of numerical or categorical data. They are commonly used in various fields, including data analysis, biology, finance etc.

```python
corr = data.corr()
plt.figure(figsize=(30,30))
sns.heatmap(corr, cmap ='Greens', annot= True)
plt.show()
```
